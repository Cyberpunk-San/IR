## 9. Outlier Detection and Treatment (Winsorization)

**Logic:** This section identifies and treats outliers in selected numerical columns using the Interquartile Range (IQR) method. Outliers are capped at the upper and lower bounds (Q1 - 1.5*IQR and Q3 + 1.5*IQR) to reduce their influence on the model, a technique known as Winsorization.

**Methods Used:**
- `df.quantile()`: Calculates the first (Q1) and third (Q3) quartiles.
- `IQR = Q3 - Q1`: Computes the Interquartile Range.
- `df.clip()`: Caps values outside the calculated IQR bounds.

```python
df['Stress Level']
```

```python
import numpy as np

num_df = df.select_dtypes(include=[np.number])

Q1 = num_df.quantile(0.25)
Q3 = num_df.quantile(0.75)
IQR = Q3 - Q1

outliers = ((num_df < (Q1 - 1.5 * IQR)) | (num_df > (Q3 + 1.5 * IQR)))

print(outliers.sum())
```

```python
outlier_rows = num_df[((num_df < (Q1 - 1.5 * IQR)) |
                       (num_df > (Q3 + 1.5 * IQR))).any(axis=1)]

print(outlier_rows)
```

```python
col = 'Physical Activity Level'

outliers_qos = df[(df[col] < (Q1[col] - 1.5 * IQR[col])) |
                  (df[col] > (Q3[col] + 1.5 * IQR[col]))]

print(outliers_qos)
```

```python
print(outliers_qos[col])
```

```python
df.info()
```

```python
outlier_cols = [
    'Sleep Duration',
    'Quality of Sleep',
    'Physical Activity Level',
    'Heart Rate',
    'Daily Steps'
]
```

```python
for col in outlier_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower, upper)
```

```python

print(df[outlier_cols].describe())
```

## 10. Feature Scaling (Min-Max Scaling)

**Logic:** The features (X) are separated from the target variable (y), 'Stress Level'. The numerical features in X are then scaled using `MinMaxScaler` to bring them within a specific range (typically 0 to 1). This is crucial for algorithms that are sensitive to the magnitude of feature values.

**Methods Used:**
- `df.drop()`: Separates features (X) and target (y).
- `MinMaxScaler`: Scales numerical features to a 0-1 range.
- `pd.DataFrame()`: Converts the scaled NumPy array back into a pandas DataFrame.

```python
X = df.drop('Stress Level', axis=1)
y = df['Stress Level']
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
import pandas as pd

X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
final_df = pd.concat([X_scaled, y], axis=1)
```
